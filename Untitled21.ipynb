```python
import sys
sys.path.append('/content/biodiffusion/src')
from ddpm_conditional import Diffusion
```

```python
# Set the number of samples and device
n = 4
device = "cuda:0"
# Create a Diffusion model instance and load the trained model checkpoint
diffusion = Diffusion(seq_size=256, num_classes=4, device=device)
diffusion.load("/content/drive/MyDrive/my_model/DDPM_conditional")
# Prepare labels and sample from the diffusion model
labels = torch.full((n,), 1).long().to(diffusion.device)
sampled_signals = diffusion.sample(use_ema=False, labels=labels)
print(sampled_signals.shape)
```

```python
import matplotlib.pyplot as plt
def plot_3d_signals(signals):
  num_labels, num_dimensions, sequence_length, _ = signals.shape

  for label_idx in range(num_labels):
      fig, axes = plt.subplots(num_dimensions, 1, figsize=(10, 8), sharex=True)
      fig.suptitle(f'Signals for label {label_idx + 1}')

      # Extract the 3D signal for the current label
      signal = signals[label_idx]  # Shape: [num_dimensions, sequence_length, 1]

      # Plot each dimension
      for dim_idx in range(num_dimensions):
          axes[dim_idx].plot(signal[dim_idx, :, 0], label=f'Dimension {dim_idx + 1}')
          axes[dim_idx].legend()
          axes[dim_idx].grid(True)
          axes[dim_idx].set_ylabel(f'Dim {dim_idx + 1}')

      axes[-1].set_xlabel('Sequence Length')

      plt.tight_layout()
      plt.show()
```

```python
plot_3d_signals(sampled_signals.cpu().detach().numpy())
```
